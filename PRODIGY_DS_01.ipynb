**Q: Create a bar chart or histogram to visualize a distribution of a categorical or continuous variables, such as a distribution of ages or genders in a population**

```python
import matplotlib.pyplot as plt

# Sample data
genders = ['Male', 'Female', 'Non-Binary', 'Other']
counts = [450, 520, 80, 50]  # Sample counts for each gender

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(genders, counts, color=['green', 'pink', 'orange', 'grey'])

# Adding title and labels
plt.title('Distribution of Genders in Population')
plt.xlabel('Gender')
plt.ylabel('Count')

# Display the plot
plt.show()
```
